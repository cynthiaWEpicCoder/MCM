```python
import pandas as pd

#EVENT PARTICIPATION DATA
athletes = pd.read_csv('./summerOly_athletes.csv')
event_data = athletes.drop_duplicates()
#event_counts = event_data.groupby(['Year', 'NOC']).size().reset_index(name='Athlete_Count')
#event_pivot_data = event_counts.pivot(index='Year', columns='NOC', values='Athlete_Count')
events_summary = event_data.groupby(['Team', 'Year'])['Event'].nunique().reset_index(name='Event_Count')

events_summary

#df = pd.read_csv(url, names=['Event Participation', 'Past Medal Count', 'Host Country Advantage', 'Cluster 1 Specialization','Cluster 1 Specialization',])
```

```python
#PAST MEDAL COUNT
medals_summary = event_data.groupby(['Team', 'Year']).size().reset_index(name='Medal_Count')
medals_summary
```

```python
# Create a complete list of all combinations of Team and Year
all_combinations = event_data[['Team', 'Year']].drop_duplicates()

# Merge with medal counts and fill missing values with 0
complete_medals_summary = all_combinations.merge(medals_summary, on=['Team', 'Year'], how='left').fillna({'Medal_Count': 0})
print(medals_summary)

# Merge with event counts and fill missing values with 0
complete_events_summary = all_combinations.merge(events_summary, on=['Team', 'Year'], how='left').fillna({'Event_Count': 0})

# Convert counts to integers (optional, for cleaner representation)
complete_medals_summary['Medal_Count'] = complete_medals_summary['Medal_Count'].astype(int)
complete_events_summary['Event_Count'] = complete_events_summary['Event_Count'].astype(int)

# Display the first few rows of the updated DataFrames
complete_medals_summary.head(30), complete_events_summary
```
